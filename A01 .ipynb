## Contador multiplos de 3

```python
lista100 = []
```

```python
def multiplos (valor, multiplo):
    return True if valor % multiplo == 0 else False
```

```python
for x in range (1,100):
    if multiplos(x,3):
        lista100.append(x)
```

## DatosCT

### Utilidades

```python
datosCT = pd.read_excel('datosCT.xlsx')
datosCT['Utilidades']=datosCT['Ingreso']-datosCT['Gasto']
```

```python
datosCT['Ingreso'].sum()
```

```python
datosCT['Gastos en pérdidas'] = datosCT['Gasto'].where(datosCT['Utilidades']<0,0)
```

```python
datosCT
```
